```python
import os
import numpy as np

def do_split(arr):
    arr=arr.split('[')
    # print(arr[1])
    arr=arr[1].split(']')
    arr=arr[0]
    arr=arr.split('\'')

    re=', '
    re1='\''
    re2=''
    while re in arr:
        ind=arr.index(re)
        arr.pop(ind)

    while re1 in arr:
        ind=arr.index(re1)
        arr.pop(ind)
    while re2 in arr:
        ind=arr.index(re2)
        arr.pop(ind)
    # print(arr)
    brr=[]
    for j in range(len(arr)):
        # print(arr[j])
        if arr[j]==', 0, ':
            brr.append(0)
            brr.append(0)
            brr.append(0)
        else:
            crr=arr[j].split(',')
            # print(crr)
            for i in crr:
                if i=='' or i==' ':
                    continue
        
                temp=i.split('(')
                if len(temp)>1:
                    # print(temp[1])
                    brr.append(float(temp[1]))
                    continue
                temp=i.split(')')
                if len(temp)>1:
                    brr.append(float(temp[0]))
                    continue
                # print(i)
        
                if i==' 0' or i=='0':
                    brr.append(0)
                    brr.append(0)
                    brr.append(0)
                    continue
                brr.append(float(i))
    # print(brr)
    # print(len(brr))
    if len(brr)!=36:
        print("length is not 36!!")
    return brr
	


```

```python
n_timesteps, n_features, n_outputs = trainX.shape[1], trainX.shape[2], testy.shape[1]
print("Time Steps : ", n_timesteps)
print("#Features : ", n_features)
print("#Outputs : ", n_outputs)
```

```python
from keras.models import Sequential
from keras.layers.recurrent import LSTM
from keras.layers.convolutional_recurrent import ConvLSTM2D
from keras.layers import Dense, Softmax, Dropout
```

```python
def evaluate_model(trainX, trainy, testX, testy):
    verbose, epochs, batch_size = 1, 50,32
    n_timesteps, n_outputs = trainX.shape[1], testy.shape[1]
    model = Sequential()
    model.add(LSTM(120, input_shape=(n_timesteps,n_features)))
    #model.add(ConvLSTM2D(100, kernel_size=(1,2), input_shape=(None,None,n_timesteps,n_features)))
    model.add(Dropout(0.3))
    model.add(Dense(100, activation='relu'))
    model.add(Dense(n_outputs, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    # fit network
    model.fit(trainX, trainy, epochs=epochs, batch_size=batch_size, verbose=verbose)
    # evaluate model
    _, accuracy = model.evaluate(testX, testy, batch_size=batch_size, verbose=1)
    return model
```
